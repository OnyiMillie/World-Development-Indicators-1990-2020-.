```python
# GDPCategory = "Low" / "Medium" / "High" based on:
○	Low: below median GDP
○	Medium: between median and 75th percentile
○	High: above 75th percentile
```

```python
# Sample DataFrame
data = {
    "Country": ["USA", "Canada", "Mexico", "Brazil", "France", "Germany", "Japan", "UK", "Italy", "Spain", "Australia", "Norway"],
    "GDPperCapita": [65000, 45000, 15000, 12000, 42000, 48000, 40000, 41000, 35000, 33000, 52000, 60000]
}
df = pd.DataFrame(data)
```

```python
# Compute median and 75th percentile
median_gdp = df["GDPperCapita"].median()
percentile_75 = df["GDPperCapita"].quantile(0.75)
```

```python
# Create GDPCategory column
def categorize_gdp(value):
    if value < median_gdp:
        return "Low"
    elif median_gdp <= value <= percentile_75:
        return "Medium"
    else:
        return "High"
```

```python
df["GDPCategory"] = df["GDPperCapita"].apply(categorize_gdp)

# Count how many rows fall into each category
category_counts = df["GDPCategory"].value_counts()

print("Number of rows per GDP category:\n")
print(category_counts)
```

```python
Only three countries falls within the High and Medium categories respectively, while 6 falls within the Low category
```

# TASK 10 – Compare 1990 vs 2020

```python
# Identify countries with data for both 1990 and 2020
```

```python
# Sample dataset
data = {
    "Country": ["USA", "Canada", "Mexico", "Brazil", "France", "Germany", "Japan", "UK", "USA", "Canada", "Mexico", "Brazil"],
    "Year": [1990, 1990, 1990, 1990, 1990, 1990, 1990, 1990, 2020, 2020, 2020, 2020],
    "LifeExpectancy": [75, 77, 70, 65, 78, 76, 80, 76, 79, 82, 76, 74]
}
df = pd.DataFrame(data)
```

```python
# Identify countries with data for both 1990 and 2020
countries_with_both = df.groupby("Country")["Year"].apply(lambda x: set([1990, 2020]).issubset(set(x)))

# Filter only those countries
result = countries_with_both[countries_with_both].index.tolist()

print("Countries with data for both 1990 and 2020:\n")
print(result)
```

```python
Countries with 1990 and 2020 data were 'Brazil', 'Canada', 'Mexico', 'USA'
```

```python
# Compute change in LifeExpectancy for those countries
```

```python
# Filter only the countries of interest
countries = ["Brazil", "Canada", "Mexico", "USA"]
df_subset = df[df["Country"].isin(countries)]

# Pivot to have years as columns
pivot_df = df_subset.pivot(index="Country", columns="Year", values="LifeExpectancy")
```

```python
# Compute change between 1990 and 2020
pivot_df["Change_LifeExpectancy"] = pivot_df[2020] - pivot_df[1990]

print("Change in Life Expectancy (1990 → 2020):\n")
print(pivot_df[["Change_LifeExpectancy"]])
```

```python
The change(increase) in life expectancy of 'Brazil', 'Canada', 'Mexico', 'USA' between 1990 and 2020 were 9, 5, 6, and 4 respectively.
```

```python
# List the top 3 countries with the greatest improvement
```

```python
# Sample dataset with Life Expectancy in two years
data = {
    "Country": ["USA", "Canada", "Mexico", "Brazil", "France", "Germany", "Japan", "UK"],
    "LifeExp_2000": [77, 79, 72, 68, 80, 78, 81, 78],
    "LifeExp_2020": [79, 82, 76, 74, 83, 81, 85, 81]
}
df = pd.DataFrame(data)
```

```python
# Compute improvement
df["Improvement"] = df["LifeExp_2020"] - df["LifeExp_2000"]

# Sort by improvement and select top 3
top3 = df.sort_values(by="Improvement", ascending=False).head(3)

print("Top 3 countries with greatest improvement in Life Expectancy:\n")
print(top3[["Country", "Improvement"]])
```

```python
The top 3 countries with the greatest improvement were Brazil, Mexico and Japan with improvement 6 and 4 respectively.
```
